### Save data from test 2 dataset (NASA)

```python
import numpy as np
import scipy.io as sio
import pandas as pd
from os import listdir
```

```python
signals_Ch1 = []
signals_Ch2 = []
signals_Ch3 = []
signals_Ch4 = []
n_files = len(listdir("../2nd_test"))
cont = 0
for archivo in listdir("../2nd_test"):
    archivo = '../2nd_test/' + archivo
    df = pd.read_csv(archivo, sep = '\t', header = None)
    df.columns = ['Ch1', 'Ch2', 'Ch3', 'Ch4']
    cont += 1
    print(str(cont) + ' files read out of ' + str(n_files))
    signals_Ch1.append(df.Ch1.values)
    signals_Ch2.append(df.Ch2.values)
    signals_Ch3.append(df.Ch3.values)
    signals_Ch4.append(df.Ch4.values)
```

```python
signals_test2 = pd.DataFrame({'Ch1': signals_Ch1, 'Ch2': signals_Ch2, 'Ch3': signals_Ch3, 'Ch4': signals_Ch4})
signals_test2
```

#### We group all data by channel:

```python
canal1 = signals_test2['Ch1']
canal1 = np.concatenate(canal1[0:])
```

```python
canal2 = signals_test2['Ch2']
canal2 = np.concatenate(canal2[0:])
```

```python
canal3 = signals_test2['Ch3']
canal3 = np.concatenate(canal3[0:])
```

```python
canal4 = signals_test2['Ch4']
canal4 = np.concatenate(canal4[0:])
```

```python
sio.savemat('./DataSignals/signals_test2_channels.mat', {'Ch1': canal1, 'Ch2': canal2, 'Ch3': canal3, 'Ch4': canal4})
```
