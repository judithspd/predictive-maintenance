### Visualization of the 4 channels of test 3 NASA Bearing dataset.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import scipy.io as sio
```

```python
signals_test3_channels = sio.loadmat('./DataSignals/signals_test3_channels.mat')
signals_test3_channels
```

```python
canal1 = signals_test3_channels['Ch1']
canal2 = signals_test3_channels['Ch2']
canal3 = signals_test3_channels['Ch3']
canal4 = signals_test3_channels['Ch4']
canal1 = canal1[0]
canal2 = canal2[0]
canal3 = canal3[0]
canal4 = canal4[0]
t =  np.linspace(0, 10 * len(canal1), len(canal1))
```

```python
len(canal1)
```

```python
plt.figure(figsize=(7,4))
plt.plot(t, canal1)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Ch1 - test3')
```

```python
plt.figure(figsize=(7,4))
plt.plot(t, canal2)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Ch2 - test3')
```

```python
plt.figure(figsize=(7,4))
plt.plot(t, canal3)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Ch3 - test3')
```

```python
plt.figure(figsize=(7,4))
plt.plot(t, canal4)
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Ch4 - test3')
```
