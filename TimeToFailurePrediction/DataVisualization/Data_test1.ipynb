### Save data from test 2 dataset (NASA)

```python
import numpy as np
import scipy.io as sio
import pandas as pd
from os import listdir
```

```python
signals_Ch1 = []
signals_Ch2 = []
signals_Ch3 = []
signals_Ch4 = []
signals_Ch5 = []
signals_Ch6 = []
n_files = len(listdir("../1st_test"))
cont = 0
for file in listdir("../1st_test"):
    file = '../1st_test/' + file
    df = pd.read_csv(file, sep = '\t', header = None)
    df.columns = ['Ch1', 'Ch2', 'Ch3', 'Ch4', 'Ch5', 'Ch6', 'Ch7', 'Ch8']
    cont += 1
    print(str(cont) + ' files read out of ' + str(n_files))
    signals_Ch1.append(df.Ch1.values)
    signals_Ch2.append(df.Ch2.values)
    signals_Ch3.append(df.Ch3.values)
    signals_Ch4.append(df.Ch4.values)
    signals_Ch5.append(df.Ch5.values)
    signals_Ch6.append(df.Ch6.values)
```

```python
signals_test1 = pd.DataFrame({'Ch1': signals_Ch1, 'Ch2': signals_Ch2, 'Ch3': signals_Ch3, 'Ch4': signals_Ch4,
                              'Ch5': signals_Ch5, 'Ch6': signals_Ch6})
signals_test1
```

#### We group all data by channel:

```python
sio.savemat('./DataSignals/signals_test1_channels.mat', {'Ch1': canal1, 'Ch2': canal2, 'Ch3': canal3, 'Ch4': canal4, 
                                           'Ch5': canal5, 'Ch6': canal6})
```
