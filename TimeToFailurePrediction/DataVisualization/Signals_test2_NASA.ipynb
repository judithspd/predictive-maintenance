### Visualization of the 4 channels of test 2 NASA Bearing dataset.

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import scipy.io as sio
```

```python
signals_test2_channels = sio.loadmat('./DataSignals/signals_test2_channels.mat')
signals_test2_channels
```

```python
canal1 = signals_test2_channels['Ch1']
canal2 = signals_test2_channels['Ch2']
canal3 = signals_test2_channels['Ch3']
canal4 = signals_test2_channels['Ch4']
canal1 = canal1[0]
canal2 = canal2[0]
canal3 = canal3[0]
canal4 = canal4[0]
t =  np.linspace(0, 10 * len(canal1), len(canal1))
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.set_size_inches(15,12)
ax1.plot(t, canal1)
ax1.set_xlabel('Time')
ax1.set_ylabel('Amplitude')
ax1.set_title('Ch1 - test2')

ax2.plot(t, canal2)
ax2.set_xlabel('Time')
ax2.set_ylabel('Amplitude')
ax2.set_title('Ch2 - test2')

ax3.plot(t, canal3)
ax3.set_xlabel('Time')
ax3.set_ylabel('Amplitude')
ax3.set_title('Ch3 - test2')

ax4.plot(t, canal4)
ax4.set_xlabel('Time')
ax4.set_ylabel('Amplitude')
ax4.set_title('Ch4 - test2')
```
