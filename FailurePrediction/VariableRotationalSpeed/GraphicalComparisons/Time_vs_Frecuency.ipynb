## Time vs Frecuency

```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
import statistics as stats
import pandas as pd
from scipy.fft import fft, fftfreq, fftshift

HA1 = sio.loadmat('H-A-1.mat')
channel1HA1 = HA1['Channel_1']
n = len(channel1HA1) # NÃºmero de muestras

canal1HA1 = np.zeros(n)
for i in range(0,n):
    canal1HA1[i] = channel1HA1[i][0]

f = 200000 # Hz #graba a 200000 por s
t = np.linspace(0,9,n)
```

```python
fig = plt.figure(figsize=(9,6))
plt.plot(t, canal1HA1)
plt.ylabel('Amplitude', fontsize = 12)
plt.xlabel('Time', fontsize = 12)
plt.title('Time spectrum of signal $x$.', fontsize = 14)
```

```python
dt = t[1] - t[0]
n = len(canal1HA1)
X = fftshift(fft(canal1HA1)/n)
frec = fftshift(fftfreq(n, dt))
fig = plt.figure(figsize=(9,6))
plt.plot(frec, abs(X))
plt.ylabel('|FFT(x)/N|', fontsize = 12)
plt.xlabel('Frequency', fontsize = 12)
plt.title('Frequency spectrum of signal $x$.', fontsize = 14)
```
