## Statistics pairplot

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
dataframe = pd.read_csv("./ProcessedData/statistics_10_hamming.csv" , sep = ',')
```

```python
dataframe['Tipo'][dataframe['Tipo'].values == 'Sano'] = 'Healthy'
dataframe.rename(columns={'Tipo':'Type'},
               inplace=True)
```

```python
g = sns.pairplot(dataframe, hue = "Type", markers=["o", "s", "D"], corner = True)
g._legend.remove()
handles = g._legend_data.values()
labels = g._legend_data.keys()
g.fig.subplots_adjust(top=0.95, bottom=0.05)
g.fig.legend(handles=handles, labels=labels, loc='lower center', ncol=3, prop={'size':15})
```

```python
g.savefig('statistics_pairplot', format='svg', dpi=1200)
```
