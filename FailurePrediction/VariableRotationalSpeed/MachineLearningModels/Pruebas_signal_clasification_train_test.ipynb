## Module *signal_clasification.py* testing


Given the 36 signals of https://www.sciencedirect.com/science/article/pii/S2352340918314124, which had a duration of 10s each, these have been decomposed into 360 signals of 1s each, which are stored in the file signal1s.mat, in the variable 'Signal'. In addition, this file also stores the corresponding signal type: healthy, with inner race failure or with outer race failure (in the variable 'Tipo'). 

In this script, we are going to classify each of these 360 signals using the *classification* function of the *signal_clasification.py* module. We will find the number of classification failures (we can do this since we know the type of each signal) and measure the total time to classify all the test and train signals.

```python
import numpy as np
import scipy.io as sio
import pandas as pd
import random
import seaborn as sns
import signal_clasification_traintest
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from signal_clasification_traintest import *
from time import time
```

```python
signals1s_test = sio.loadmat('signals_test.mat')
signals_test = signals1s_test['Signal']
tipo_test = signals1s_test['Tipo']
tipo_test = np.core.defchararray.replace(tipo_test, ' ', '')
```

```python
signals1s_train = sio.loadmat('signals_train.mat')
signals_train = signals1s_train['Signal']
tipo_train = signals1s_train['Tipo']
tipo_train = np.core.defchararray.replace(tipo_train, ' ', '')
```

```python
t0 = time()
pred_test = []
tipo_test_10 = []
for i in range(len(tipo_test)):
    for j in range(10):
        signal = signals_test[i][200000 * j : 200000 * (j+1)]
        tipo_test_10.append(tipo_test[i])
        prediction = clasification(signal)
        pred_test.append(prediction)
        # print(prediction)
tiempo_total = time() - t0
print("Elapsed time: %0.3fs"%(tiempo_total))
```

```python
t0 = time()
pred_train = []
tipo_train_10 = []
for i in range(len(tipo_train)):
    for j in range(10):
        signal = signals_train[i][200000 * j : 200000 * (j+1)]
        tipo_train_10.append(tipo_train[i])
        prediction = clasification(signal)
        pred_train.append(prediction)
        # print(prediction)
tiempo_total = time() - t0
print("Elapsed time: %0.3fs"%(tiempo_total))
```

```python
print(confusion_matrix(tipo_test_10, pred_test))
print(classification_report(tipo_test_10, pred_test))
```

```python
sns.set()
mat = confusion_matrix(tipo_test_10, pred_test)
fig, ax = plt.subplots(figsize=(7,6))
sns.set(font_scale=1.3)
sns.heatmap(mat.T, square=False, annot=True, fmt='d', cbar=False,
            xticklabels=['Fallo inner race', 'Fallo oute race', 'Healthy'],
            yticklabels=['Fallo inner race', 'Fallo oute race', 'Healthy'],
            cmap=sns.cubehelix_palette(light=1, as_cmap=True))

plt.xlabel('true label');
plt.ylabel('predicted label');
```

```python
fig.savefig('majority_test.png')
```

```python
print(confusion_matrix(tipo_train_10, pred_train))
print(classification_report(tipo_train_10, pred_train))
```

```python
sns.set()
mat = confusion_matrix(tipo_train_10, pred_train)
fig, ax = plt.subplots(figsize=(7,6))
sns.set(font_scale=1.3)
sns.heatmap(mat.T, square=False, annot=True, fmt='d', cbar=False,
            xticklabels=['Fallo inner race', 'Fallo oute race', 'Healthy'],
            yticklabels=['Fallo inner race', 'Fallo oute race', 'Healthy'],
            cmap=sns.cubehelix_palette(light=1, as_cmap=True))

plt.xlabel('true label');
plt.ylabel('predicted label');
```

```python
'''
If the value of a cell is 'Sano', it sets the background color to LightGreen, IndianRed if it is 'Inner', 
and LightSalmon if it is 'Outer'.
'''
def text_format(val):
    color = 'white'
    if val == 'Sano':
        color = 'LightGreen' 
    elif val == 'Inner':
        color = 'IndianRed'
    elif val == 'Outer':
        color = 'LightSalmon'
    return 'background-color: %s' % color
```
