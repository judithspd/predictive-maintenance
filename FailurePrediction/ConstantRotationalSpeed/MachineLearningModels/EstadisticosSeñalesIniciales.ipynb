```python
import scipy.io as sio
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import lee_dataset_CWRU
from lee_dataset_CWRU import *
import estadisticos_hamming
from estadisticos_hamming import *
```

```python
faultRates = [3.585, 5.415, 1] #[cage, ball, outer, inner, shaft]
Fs = 12000
DE_H1, FE_H1, t_DE_H1, t_FE_H1, RPM_H1, samples_s_DE_H1, samples_s_FE_H1 = lee_dataset('../DataCWRU/97.mat')
DE_H2, FE_H2, t_DE_H2, t_FE_H2, RPM_H2, samples_s_DE_H2, samples_s_FE_H2 = lee_dataset('../DataCWRU/98.mat')
RPM_H2 = 1778
DE_H3, FE_H3, t_DE_H3, t_FE_H3, RPM_H3, samples_s_DE_H3, samples_s_FE_H3 = lee_dataset('../DataCWRU/99.mat')
RPM_H3 = 1750
DE_H4, FE_H4, t_DE_H4, t_FE_H4, RPM_H4, samples_s_DE_H4, samples_s_FE_H4 = lee_dataset('../DataCWRU/100.mat')
```

```python
kurt_sano = []
impulse_factor_sano = []
RMS_sano = [] 
margin_factor_sano = []
skewness_sano = []
shape_factor_sano = []
peak_to_peak_sano = []
crest_factor_sano = []
signals = [DE_H1, DE_H2, DE_H3, DE_H4]
for i in range(len(signals)):
    signal = signals[i]
    kurt, impulse_factor, RMS, margin_factor, skewness, shape_factor, peak_to_peak, crest_factor = parameters_hamming(signal)
    kurt_sano.append(kurt)
    impulse_factor_sano.append(impulse_factor)
    RMS_sano.append(RMS)
    margin_factor_sano.append(margin_factor)
    skewness_sano.append(skewness)
    shape_factor_sano.append(shape_factor)
    peak_to_peak_sano.append(peak_to_peak)
    crest_factor_sano.append(crest_factor)
```

```python
df_sano = pd.DataFrame({'Tipo': ['Sano'] * len(kurt_sano), 'Kurtosis': kurt_sano,
                                   'Impulse factor': impulse_factor_sano, 
                                   'RMS': RMS_sano, 'Margin factor': margin_factor_sano, 'Skewness': skewness_sano,
                                   'Shape factor': shape_factor_sano, 'Peak to peak': peak_to_peak_sano,
                                   'Crest factor': crest_factor_sano})

df_sano
```
