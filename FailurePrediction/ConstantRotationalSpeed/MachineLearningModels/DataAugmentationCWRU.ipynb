### New data are generated from CWRU signals.

```python
import pandas as pd
import numpy as np
from numpy.random import normal
```

```python
df_completo = pd.read_csv("./Data/EstadisticosSeñalesInicialesCWRU.csv" , sep = ',')
```

```python
X = df_completo[['Kurtosis', 'Impulse factor', 'RMS', 'Margin factor', 'Skewness',
               'Shape factor', 'Peak to peak', 'Crest factor']].values
y = df_completo['Tipo'].values
```

```python
for i in range(len(df_completo)): 
    print(df_completo.loc[i].values[0])
    print(df_completo.loc[i].values[1:])
```

```python
datos_aumentados = []
y_aumentados = []
for i in range(len(df_completo)): 
    fila = df_completo.loc[i].values[1:]
    datos_aumentados.append(fila)
    y_aumentados.append(df_completo.loc[i].values[0])
for i in range(len(df_completo)): 
    for _ in range(10):
        fila = df_completo.loc[i].values[1:]
        gauss = normal(loc=0.0, scale=0.2, size=len(fila))
        fila = fila + gauss
        datos_aumentados.append(fila)
        y_aumentados.append(df_completo.loc[i].values[0])
```

```python
estadisticos = ['Kurtosis', 'Impulse factor', 'RMS', 'Margin factor', 'Skewness',
               'Shape factor', 'Peak to peak', 'Crest factor']
df_aumentado = pd.DataFrame(datos_aumentados, columns = estadisticos)
df_aumentado.insert(0, 'Tipo', y_aumentados)
df_aumentado
```

```python
df_aumentado.to_csv('./Data/estadisticosDatosAumentadosCWRU.csv', index = False, header = True, sep = ',')
```
